# A forced dynamical system

The Lorenz attractor forced by the Rossler attractor. The Lorenz response measurements have been filtered with random Gaussian kernels

```python
X_measure = np.load("./benchmarks/data/lorenz_data.npy", allow_pickle=True)
y_driver = np.load("./benchmarks/data/lorenz_driver.npy", allow_pickle=True)
```

```python

standardize_ts = lambda ts: (ts - np.mean(ts)) / np.std(ts)

plt.figure(figsize=(10, 12))
plt.plot(10 + standardize_ts(y_driver), linewidth=3, color='b')
# plt.title("True Driver")


for i in range(50):
    plt.plot(-i + standardize_ts(X_measure[:, i]), 'k', linewidth=1, alpha=0.5);

# remove frame and ticks
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.gca().axis('off')
plt.title("Driving signal and response time series")
```

```python
from shrec.models import RecurrenceManifold
model = RecurrenceManifold(random_state=0)
y_reconstructed = model.fit_predict(X_measure)

plt.figure()
plt.plot(y_driver)

plt.figure()
plt.plot(y_reconstructed)
```

```python
# model.adjacency_matrix
# https://dyeun.wordpress.ncsu.edu/files/2022/08/Sigmetrics20-FV-1.pdf

## Simulate a diffusion process on a similarity graph

n_walkers = 500
A = model.adjacency_matrix.copy()

# Normalize A to make it a stochastic matrix
row_sums = A.sum(axis=1)
stochastic_matrix = A / row_sums[:, np.newaxis]

# Initialize
num_nodes = stochastic_matrix.shape[0]
num_steps = 2000 # Define the number of steps in the random walk
walker_distribution = np.full(num_nodes, 1/num_nodes) # Uniform distribution

# Simulate the random walk
for _ in range(num_steps):
    walker_distribution = stochastic_matrix.T @ walker_distribution

# The walker_distribution now represents the steady state distribution
# print(walker_distribution)

```

```python
import scipy.linalg
eigvals, eigvecs = scipy.linalg.eigh(
    model.adjacency_matrix, 
    subset_by_index=[model.adjacency_matrix.shape[0]-2, model.adjacency_matrix.shape[0]-1]
)

plt.figure()
plt.plot(-standardize_ts(eigvecs[:, -1]))
plt.plot(standardize_ts(walker_distribution))

from sklearn.manifold import SpectralEmbedding

embedder = SpectralEmbedding(
    n_components=1, 
    random_state=0, 
    affinity='precomputed'
)
driver_estimate = embedder.fit_transform(model.adjacency_matrix).squeeze()

plt.figure()
plt.plot(-standardize_ts(eigvecs[:, -2]))
plt.plot(standardize_ts(driver_estimate))
```
