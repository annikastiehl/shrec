```python
class SpectralEmbedding:

    def __init__(self, n_components=2, random_state=0, affinity='precomputed'):
        self.n_components = n_components
        self.random_state = random_state
        self.affinity = affinity

    def fit_transform(self, X):
        if self.affinity == 'precomputed':
            a = X
        else:
            raise NotImplementedError()
        eigvals, eigvecs = scipy.linalg.eigh(a, subset_by_index=[a.shape[0]-self.n_components, a.shape[0]-1])
        return eigvecs

    def fit(self, X):
        self.embedding_ = self.fit_transform(X)
        return self

    def transform(self, X):
        return self.embedding_

    def fit_predict(self, X):
        return self.fit_transform(X)

    def predict(self, X):
        return self.transform(X)

        
```

```python
def make_graph_laplacian(a):
    """
    Make the graph laplacian for a given adjacency matrix
    """
    d = np.sum(a, axis=0)
    d = np.diag(d)
    return d - a

import numpy as np
from hdbscan import HDBSCAN
from sklearn.cluster import KMeans

def spectral_clustering(affinity_matrix, num_clusters):

    assert np.allclose(affinity_matrix, affinity_matrix.T), "Affinity matrix must be symmetric"

    # Compute the degree matrix, which is a diagonal matrix where the
    # i-th element is the sum of the affinity matrix's i-th row
    degree_matrix = np.sum(affinity_matrix, axis=1)

    # Compute the Laplacian matrix, which is the degree matrix minus
    # the affinity matrix
    laplacian_matrix = degree_matrix - affinity_matrix

    # Compute the eigenvectors of the Laplacian matrix
    eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)

    # Sort the eigenvectors by their corresponding eigenvalues
    idx = eigenvalues.argsort()[::-1]
    eigenvectors = eigenvectors[:, idx]
    eigenvalues = eigenvalues[idx]

    # Select the top `num_clusters` eigenvectors
    eigenvectors = eigenvectors[:, :num_clusters]

    # Normalize the eigenvectors by dividing each eigenvector by its
    # corresponding eigenvalue
    # eigenvectors = eigenvectors / eigenvalues[:, None]

    # Use the eigenvectors to cluster the data
    # clusters = KMeans(n_clusters=num_clusters, random_state=0).fit_predict(eigenvectors)
    clusters = HDBSCAN(min_cluster_size=int(0.1 * affinity_matrix.shape[0])).fit_predict(eigenvectors)

    return clusters
```
