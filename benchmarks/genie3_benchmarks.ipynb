# Genie3 regulatory network inference

We start by installing dependencies and the Genie3 package via BioConductor

```r
if (!require("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

# The following initializes usage of Bioc devel
BiocManager::install(version='devel', ask = FALSE)
BiocManager::install("GENIE3")

# Install numpy utilities
if (!require("reticulate")) {
  install.packages("reticulate")
}
```

```r
library(reticulate)
np <- import("numpy")
# data reading
exprMatr <- np$load("X_measure_72.npy",  allow_pickle = TRUE)
# Check if it's an array and then transpose
exprMatr <- as.data.frame(exprMatr)
exprMatr <- t(exprMatr)

rownames(exprMatr) <- paste("Gene", 0:11, sep="")
colnames(exprMatr) <- paste("Sample", 0:499, sep="")

library(GENIE3)
set.seed(0) # For reproducibility of results
weightMat <- GENIE3(exprMatr)

# exprMatr <- matrix(sample(1:10, 100, replace=TRUE), nrow=20)
# rownames(exprMatr) <- paste("Gene", 0:19, sep="")
# colnames(exprMatr) <- paste("Sample", 0:4, sep="")
# head(exprMatr)

# library(GENIE3)
# set.seed(0) # For reproducibility of results
# weightMat <- GENIE3(exprMatr)
```

```r
# Sort nodes by out-degree in descending order
out_degree <- rowSums(weightMat)
sorted_nodes <- order(out_degree, decreasing = TRUE)
# List nodes and their corresponding out-degrees
out_degree_list <- data.frame(Node = sorted_nodes, Out_Degree = out_degree[sorted_nodes])

# Display the result
print(out_degree_list)

# Get the node indices sorted by descending order of out-degree
sorted_nodes <- order(out_degree, decreasing = TRUE)

# Adjust to 0-based indexing
sorted_nodes_zero_based <- sorted_nodes - 1
print(sorted_nodes_zero_based)
```

```r
# drive_ls()

library(googledrive)
all_files <- drive_ls(path = "combination_regulation/")
# Define the pattern to match 'X' followed by anything, ending with '.npy'
pattern <- "^X.*\\.npy$"
# Filter files that match the pattern
npy_files <- all_files[grep(pattern, all_files$name),]
```

```r
# Iterate through the files and download them locally
for (i in 1:nrow(npy_files)) {
  file_id <- npy_files$id[i]
  file_name <- npy_files$name[i]

  # Download each file to local environment in Colab
  drive_download(as_id(file_id), path = paste0("/content/", file_name), overwrite = TRUE)
}
```

```r
# # Define the directory containing .npy files
directory <- "./"

# Get a list of all .npy files in the directory
npy_files <- list.files(path = directory, pattern = "\\.npy$", full.names = TRUE)
print(length(npy_files))
# npy_files <- drive_ls(path = "combination_regulation/")
# print(npy_files)

# Loop through each file
for (file in npy_files) {
# for (i in seq_len(nrow(npy_files))) {
#   # Access file information using indexing since npy_files is a data frame
#   local_file <- paste0(tempdir(), "/", npy_files[i, "name"])  # Save to a temporary directory
#   # Download the file locally
#   drive_download(as_id(npy_files[i, "id"]), path = local_file, overwrite = TRUE)


  # Data reading
  exprMatr <- np$load(file, allow_pickle = TRUE)
  exprMatr <- as.data.frame(exprMatr)
  exprMatr <- t(exprMatr)

  rownames(exprMatr) <- paste("Gene", 0:11, sep="")
  colnames(exprMatr) <- paste("Sample", 0:499, sep="")

  library(GENIE3)
  set.seed(0) # For reproducibility of results
  weightMat <- GENIE3(exprMatr)

  # Sort nodes by out-degree in descending order
  out_degree <- rowSums(weightMat)
  sorted_nodes <- order(out_degree, decreasing = TRUE)
  # List nodes and their corresponding out-degrees
  out_degree_list <- data.frame(Node = sorted_nodes, Out_Degree = out_degree[sorted_nodes])

  # Display the result
  # print(out_degree_list)

  # Get the node indices sorted by descending order of out-degree
  sorted_nodes <- order(out_degree, decreasing = TRUE)

  # Adjust to 0-based indexing
  sorted_nodes_zero_based <- sorted_nodes - 1

  number_in_filename <- sub(".*_(\\d+)\\.npy$", "\\1", file)
  print(number_in_filename)
  output_file <- paste0("output/sorted_nodes_", number_in_filename, ".txt")
  write.table(sorted_nodes_zero_based, file = output_file, row.names = FALSE, col.names = FALSE)

}
```
